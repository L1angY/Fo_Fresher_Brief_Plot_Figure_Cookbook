如果去翻找有关Python气象绘图的教程,可以发现matplotlib和cartopy是必不可少的库.
其中matplotlib是Python数据可视化过程中的根本.
想要正常的使用matplotlib,必须要搞清楚画布(Figure)和轴域(Axes)的关系.

其中Figure可以理解为一张干净的白纸,而Axes则是白纸上的一块矩形区域.
一个Figure可以含有多个Axes,而一个Axes可以包含:
曲线(line2D),坐标轴(axis),坐标轴刻度(tick),坐标轴刻度标签(tick_label),边界柱(spines),
图例(legend),文本(text),标题(title)等组件，统称为艺术家(Artist)

他们的关系可以用官方提供的图表示
![anatomy](https://matplotlib.org/stable/_images/anatomy.png)

matplotlib绘图有两种方式:面向对象和pyplot形式.
可以通过下面的代码直观的了解他们之间的差异:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 面向对象
x = np.linspace(0, 2, 100)
fig, ax = plt.subplots(figsize=(8, 6))  # 创建figure,axes
ax.plot(x, x, label='linear')  # 绘制一次函数及名称
ax.plot(x, x**2, label='quadratic')  # 绘制二次函数及名称
ax.plot(x, x**3, label='cubic')  # 绘制三次函数及名称
ax.set_xlabel('x label')  # 设置x轴名称
ax.set_ylabel('y label')  # 设置y轴名称
ax.set_title("Simple Plot")  # 设置标题
ax.legend()  # 添加图例
```

```python
# pyplot形式
plt.figure(figsize=(8, 6))
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

可以发现,不管是面向对象还是pyplot形式,本质上都是属于命令式绘图,这跟另一款著名软件NCL有很大区别.
NCL采用的是声明式绘图,NCL的代码示例如下: https://www.ncl.ucar.edu/Applications/Scripts/ce_3.ncl
其实,单纯从绘图质量来说,其实第一选项是NCL.只是Python在通用性和生态方面更为强大.

另外,但使用matplotlib绘图的需求开始复杂和精细化时,不可避免的使用使用面向对象的方式.
然而,代码中大量的set_...()方法,经常让人非常的困惑,某种程度上也失去了面向对象编程的优势
当加上cartopy进行地理数据绘图时,情况将变得更加复杂.

所以,在下面的教程中,将使用proplot库来完成绘图.
