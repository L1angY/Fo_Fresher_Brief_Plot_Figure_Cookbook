## 本教程涵盖的范围

本简明教程旨为提供提供一些简单的绘图指导,以便大家能够快速简单绘制出质量不错的Figures
本教程对操作系统没有要求,Windows/Linux/MacOS均可

---------------------------------------------

本代码暂不打算涉及Python基础语法和具体统计算法的编程指导(例如对变量场进行EOF分解),具体原因如下:
① 大部分现在在Python都可通过调用第三方库来解决
② 把所有的统计方法都介绍并讲述清楚会造成本教程非常冗余

所以,对于具体的算法代码,本代码可能会不加说明的直接使用,看不懂的可以先暂时接跳过

------------------------------------------------------

另外,相信大家都有一定的Python基础(达到能够安装第三方库的程度),因为现在小学期好像有气象Python应用的课程?
所以本教程不可能从安装配置Python开始
当然如果你没有相关基础,又打算入坑Python,我也可以告诉你我的经验

## 我的工作环境简介

① Python环境由Miniconda配置和管理,之所以使用Miniconda而不是Anaconda,是因为我需要更加自由的配置环境,并且Anaconda太过于臃肿
   当然直接Anaconda也是没有问题的.当然不管是选择Anaconda还是Miniconda,我都强烈建议不要在base环境中安装第三方库
   而是新建一个虚拟环境,并在新建的虚拟环境中安装第三方库

② 目前我是用的IDE为JetBrains公司的DataSpell(付费,可以申请edu license免费使用)
   当然也可以使用其他的IDE,如PyCharm,Spyder和VsCode等等

③ 我比较常用的第三方库有:
   1. numpy: 数组处理
   2. matplotlib: 绘图库
   3. xarray: 多维数据处理工具(主要用于读取nc文件)
   4. scipy: 科学计算库
   5. sklearn: 机器(统计)学习库
   6. cartopy: 地理绘图库
   7. proplot: 绘图库(cartopy+matplotlib的高级封装)
   8. eofs: EOF分析库

④ 主力操作系统为Windows,偶尔使用Windows Subsystem for Linux(WSL)
